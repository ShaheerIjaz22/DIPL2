# Task # 6

```python
import numpy as np
img1 = cv2.imread("./img/one.jpg")
B, G, R = cv2.split(img1)
gray = 0.299 * R + 0.587 * G + 0.114 * B
img2 = np.zeros_like(img1)
img2[:, :, 0] = gray  # Blue
img2[:, :, 1] = gray  # Green
img2[:, :, 2] = gray  # Red
img2 = img2.astype(np.uint8)
cv2.imshow('Grayscale Image', img2)
cv2.waitKey(0)
```
